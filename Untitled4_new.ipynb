```python
!pip install librosa pydub matplotlib scipy
!apt-get install ffmpeg
!pip install librosa --upgrade
!pip install --upgrade librosa
# Import necessary libraries
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
from pydub import AudioSegment, effects
from scipy.io.wavfile import write
import scipy.signal as signal

# c file m thanh
filename = "signal3.wav"
y, sr = librosa.load(filename)

# Hin th tn hiu ban u
plt.figure(figsize=(12, 4))
librosa.display.waveshow(y, sr=sr)
plt.title("Tn hiu ban u")
plt.show()

# Hm chun ha tn hiu
def normalize_audio(y):
    return y / np.max(np.abs(y)) if np.max(np.abs(y)) > 0 else y

# Thay i cao 
def pitch_shift(y, sr, n_steps):
    return librosa.effects.pitch_shift(y, sr=sr, n_steps=n_steps)

# To hiu ng formant
def formant_shift(y, shift_factor):
    y_shifted = np.interp(
        np.arange(0, len(y), shift_factor),
        np.arange(0, len(y)),
        y
    )
    return y_shifted

# Thm hiu ng reverb
def add_reverb(y, sr, decay=0.3):
    impulse = np.random.randn(int(sr * decay))  # To tn hiu ngu nhin
    # Use signal.windows.hann instead of signal.hann
    impulse = signal.windows.hann(len(impulse)) * impulse  # Lm mn tn hiu
    reverb = np.convolve(y, impulse, mode='full')[:len(y)]
    return reverb

# Chuyn i ging ni theo nhn vt
def transform_voice(y, sr, character):
    if character == "Conan":
        return pitch_shift(y, sr, n_steps=2)  # Ging trung bnh
    elif character == "Shin":
        return pitch_shift(y, sr, n_steps=6)  # Ging cao, nh nhng
    elif character == "Doraemon":
        y = pitch_shift(y, sr, n_steps=-4)  # Ging trm
        y = add_reverb(y, sr, decay=0.5)  # Vang nh
        return y
    elif character == "Suspect":
        y = pitch_shift(y, sr, n_steps=-6)  # Gim pitch  trm hn
        y = librosa.effects.time_stretch(y, rate=1.1)  # Tng tc 
        return normalize_audio(y)  # Chun ha tn hiu
        return normalize_audio(y)  # Chun ha tn hiu
    elif character == "M Shin":
        y = pitch_shift(y, sr, n_steps=5)  # Gim cao 
        y = librosa.effects.time_stretch(y, rate=1.1)  # Ko di nh
        y = add_reverb(y, sr, decay=0.3)  # Gim vang
        return normalize_audio(y)  # Chun ha tn hiu
    elif character == "Nobita":
        y = pitch_shift(y, sr, n_steps=1)  # Ging trung bnh
        y = librosa.effects.time_stretch(y, rate=0.95)  # Tng chm nh
        y = formant_shift(y, shift_factor=1.1)  # Lm ging dy hn
        return y


    else:
        raise ValueError("Khng h tr nhn vt ny.")

# Chn nhn vt  x l
character = "Conan"  # Thay i theo nhn vt mong mun
y_transformed = transform_voice(y, sr, character)

# Hin th tn hiu sau khi chuyn i ging ni
plt.figure(figsize=(12, 4))
librosa.display.waveshow(y_transformed, sr=sr)
plt.title(f"Tn hiu sau khi chuyn i ging ni: {character}")
plt.show()

# Lu file m thanh sau khi chuyn i
output_filename = f"transformed_voice_{character}.wav"
write(output_filename, sr, (y_transformed * 32767).astype(np.int16))  # Chun ha tn hiu

# S dng Pydub  hon thin m thanh
audio = AudioSegment.from_file(output_filename)
audio_with_reverb = effects.normalize(audio)  # Thm hiu ng normalize
final_output = f"final_{character}_voice.wav"
audio_with_reverb.export(final_output, format="wav")

print(f"Hon tt! File ging ni  c lu ti: {final_output}")
```

```python
!pip install noisereduce
!pip install librosa matplotlib
!pip install soundfile

import librosa
import librosa.display
import matplotlib.pyplot as plt
import noisereduce as nr
import soundfile as sf


# Thay i ng dn n tp WAV cha cuc tr chuyn gia hai ngi
filename = 'signal.wav'
y, sr = librosa.load(filename)

# Hin th tn hiu m thanh ban u trong min thi gian
plt.figure(figsize=(12, 4))
librosa.display.waveshow(y, sr=sr)
plt.title("Tn hiu ban u trong min thi gian")
plt.show()

# Kh nhiu bng thut ton spectral gating
y_denoised = nr.reduce_noise(y=y, sr=sr)

# Hin th tn hiu sau khi kh nhiu
plt.figure(figsize=(12, 4))
librosa.display.waveshow(y_denoised, sr=sr)
plt.title("Tn hiu sau khi kh nhiu trong min thi gian")
plt.show()

# V Spectrogram trc v sau khi kh nhiu
plt.figure(figsize=(12, 6))

# Spectrogram ban u
plt.subplot(2, 1, 1)
X = librosa.stft(y)
Xdb = librosa.amplitude_to_db(abs(X))
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz')
plt.colorbar(format="%+2.0f dB")
plt.title("Spectrogram ban u")

# Spectrogram sau khi kh nhiu
plt.subplot(2, 1, 2)
X_denoised = librosa.stft(y_denoised)
Xdb_denoised = librosa.amplitude_to_db(abs(X_denoised))
librosa.display.specshow(Xdb_denoised, sr=sr, x_axis='time', y_axis='hz')
plt.colorbar(format="%+2.0f dB")
plt.title("Spectrogram sau khi kh nhiu")

plt.tight_layout()
plt.show()

sf.write('signal2.wav', y_denoised, sr)
```

```python
!pip install scipy numpy

from scipy.io.wavfile import read, write
import numpy as np
from scipy.signal import firwin, lfilter, kaiserord  # Thm kaiserord vo

# c file WAV
fs, signal = read('signal2.wav')

# To b lc Bandpass
nyq_rate = fs / 2.0
low_cutoff_hz = 300.0
high_cutoff_hz = 3400.0
width = 500.0 / nyq_rate
ripple_db = 60.0

N, beta = kaiserord(ripple_db, width)
if N % 2 == 0:
    N += 1
hBP_FIR = firwin(N, [low_cutoff_hz / nyq_rate, high_cutoff_hz / nyq_rate], window=('kaiser', beta), pass_zero=False)

# p dng b lc Bandpass
filtered_signal = lfilter(hBP_FIR, 1.0, signal)

# Ghi tn hiu  lc ra file WAV
write('signal3.wav', fs, np.int16(filtered_signal))
```
