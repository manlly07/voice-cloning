```python
!pip install librosa pydub matplotlib scipy
!apt-get install ffmpeg
!pip install librosa --upgrade
!pip install --upgrade librosa
# Import necessary libraries
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
from pydub import AudioSegment, effects
from scipy.io.wavfile import write
import scipy.signal as signal

# Đọc file âm thanh
filename = "signal3.wav"
y, sr = librosa.load(filename)

# Hiển thị tín hiệu ban đầu
plt.figure(figsize=(12, 4))
librosa.display.waveshow(y, sr=sr)
plt.title("Tín hiệu ban đầu")
plt.show()

# Hàm chuẩn hóa tín hiệu
def normalize_audio(y):
    return y / np.max(np.abs(y)) if np.max(np.abs(y)) > 0 else y

# Thay đổi cao độ
def pitch_shift(y, sr, n_steps):
    return librosa.effects.pitch_shift(y, sr=sr, n_steps=n_steps)

# Tạo hiệu ứng formant
def formant_shift(y, shift_factor):
    y_shifted = np.interp(
        np.arange(0, len(y), shift_factor),
        np.arange(0, len(y)),
        y
    )
    return y_shifted

# Thêm hiệu ứng reverb
def add_reverb(y, sr, decay=0.3):
    impulse = np.random.randn(int(sr * decay))  # Tạo tín hiệu ngẫu nhiên
    # Use signal.windows.hann instead of signal.hann
    impulse = signal.windows.hann(len(impulse)) * impulse  # Làm mịn tín hiệu
    reverb = np.convolve(y, impulse, mode='full')[:len(y)]
    return reverb

# Chuyển đổi giọng nói theo nhân vật
def transform_voice(y, sr, character):
    if character == "Conan":
        return pitch_shift(y, sr, n_steps=2)  # Giọng trung bình
    elif character == "Shin":
        return pitch_shift(y, sr, n_steps=6)  # Giọng cao, nhẹ nhàng
    elif character == "Doraemon":
        y = pitch_shift(y, sr, n_steps=-4)  # Giọng trầm
        y = add_reverb(y, sr, decay=0.5)  # Vang nhẹ
        return y
    elif character == "Suspect":
        y = pitch_shift(y, sr, n_steps=-6)  # Giảm pitch để trầm hơn
        y = librosa.effects.time_stretch(y, rate=1.1)  # Tăng tốc độ
        return normalize_audio(y)  # Chuẩn hóa tín hiệu
        return normalize_audio(y)  # Chuẩn hóa tín hiệu
    elif character == "Mẹ Shin":
        y = pitch_shift(y, sr, n_steps=5)  # Giảm cao độ
        y = librosa.effects.time_stretch(y, rate=1.1)  # Kéo dài nhẹ
        y = add_reverb(y, sr, decay=0.3)  # Giảm vang
        return normalize_audio(y)  # Chuẩn hóa tín hiệu
    elif character == "Nobita":
        y = pitch_shift(y, sr, n_steps=1)  # Giọng trung bình
        y = librosa.effects.time_stretch(y, rate=0.95)  # Tăng chậm nhẹ
        y = formant_shift(y, shift_factor=1.1)  # Làm giọng dày hơn
        return y


    else:
        raise ValueError("Không hỗ trợ nhân vật này.")

# Chọn nhân vật để xử lý
character = "Conan"  # Thay đổi theo nhân vật mong muốn
y_transformed = transform_voice(y, sr, character)

# Hiển thị tín hiệu sau khi chuyển đổi giọng nói
plt.figure(figsize=(12, 4))
librosa.display.waveshow(y_transformed, sr=sr)
plt.title(f"Tín hiệu sau khi chuyển đổi giọng nói: {character}")
plt.show()

# Lưu file âm thanh sau khi chuyển đổi
output_filename = f"transformed_voice_{character}.wav"
write(output_filename, sr, (y_transformed * 32767).astype(np.int16))  # Chuẩn hóa tín hiệu

# Sử dụng Pydub để hoàn thiện âm thanh
audio = AudioSegment.from_file(output_filename)
audio_with_reverb = effects.normalize(audio)  # Thêm hiệu ứng normalize
final_output = f"final_{character}_voice.wav"
audio_with_reverb.export(final_output, format="wav")

print(f"Hoàn tất! File giọng nói đã được lưu tại: {final_output}")
```

```python
!pip install noisereduce
!pip install librosa matplotlib
!pip install soundfile

import librosa
import librosa.display
import matplotlib.pyplot as plt
import noisereduce as nr
import soundfile as sf


# Thay đổi đường dẫn đến tệp WAV chứa cuộc trò chuyện giữa hai người
filename = 'signal.wav'
y, sr = librosa.load(filename)

# Hiển thị tín hiệu âm thanh ban đầu trong miền thời gian
plt.figure(figsize=(12, 4))
librosa.display.waveshow(y, sr=sr)
plt.title("Tín hiệu ban đầu trong miền thời gian")
plt.show()

# Khử nhiễu bằng thuật toán spectral gating
y_denoised = nr.reduce_noise(y=y, sr=sr)

# Hiển thị tín hiệu sau khi khử nhiễu
plt.figure(figsize=(12, 4))
librosa.display.waveshow(y_denoised, sr=sr)
plt.title("Tín hiệu sau khi khử nhiễu trong miền thời gian")
plt.show()

# Vẽ Spectrogram trước và sau khi khử nhiễu
plt.figure(figsize=(12, 6))

# Spectrogram ban đầu
plt.subplot(2, 1, 1)
X = librosa.stft(y)
Xdb = librosa.amplitude_to_db(abs(X))
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz')
plt.colorbar(format="%+2.0f dB")
plt.title("Spectrogram ban đầu")

# Spectrogram sau khi khử nhiễu
plt.subplot(2, 1, 2)
X_denoised = librosa.stft(y_denoised)
Xdb_denoised = librosa.amplitude_to_db(abs(X_denoised))
librosa.display.specshow(Xdb_denoised, sr=sr, x_axis='time', y_axis='hz')
plt.colorbar(format="%+2.0f dB")
plt.title("Spectrogram sau khi khử nhiễu")

plt.tight_layout()
plt.show()

sf.write('signal2.wav', y_denoised, sr)
```

```python
!pip install scipy numpy

from scipy.io.wavfile import read, write
import numpy as np
from scipy.signal import firwin, lfilter, kaiserord  # Thêm kaiserord vào

# Đọc file WAV
fs, signal = read('signal2.wav')

# Tạo bộ lọc Bandpass
nyq_rate = fs / 2.0
low_cutoff_hz = 300.0
high_cutoff_hz = 3400.0
width = 500.0 / nyq_rate
ripple_db = 60.0

N, beta = kaiserord(ripple_db, width)
if N % 2 == 0:
    N += 1
hBP_FIR = firwin(N, [low_cutoff_hz / nyq_rate, high_cutoff_hz / nyq_rate], window=('kaiser', beta), pass_zero=False)

# Áp dụng bộ lọc Bandpass
filtered_signal = lfilter(hBP_FIR, 1.0, signal)

# Ghi tín hiệu đã lọc ra file WAV
write('signal3.wav', fs, np.int16(filtered_signal))
```
